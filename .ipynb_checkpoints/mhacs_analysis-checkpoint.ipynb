```python
import pandas as pd

df = pd.read_csv("mhacs.csv", low_memory=False)

df.head()
```

```python
df.columns.tolist()
```

```python
selected_cols = [
    'GEN_08B',
    'DHHGAGE',
    'GENDER',
    'DHHGMS',
    'GEODVPSZ',
    'GEN_07',
    'GEN_02B',
    'SCRDDEP',
    'SCRDGAD'
]

df_selected = df[selected_cols]

df_selected.head()
```

```python
df_selected.info()
```

```python
for col in df_selected.columns:
    print(col)
    print(df_selected[col].value_counts(dropna=False))
    print("------")
```

```python
df_clean = df_selected.replace([7, 8, 9, 96, 97, 98, 99], None)
```

```python
df_clean = df_clean.dropna()
```

```python
df_clean.columns = [
    'mental_health',
    'age',
    'gender',
    'marital_status',
    'region',
    'stress',
    'life_satisfaction',
    'depression',
    'anxiety'
]
```

```python
df_clean['mental_health_binary'] = df_clean['mental_health'].apply(
    lambda x: 1 if x in [1, 2, 3] else 0
)
```

```python
import pandas as pd

table = pd.crosstab(df_clean['stress'], df_clean['mental_health_binary'])
table
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square:", chi2)
print("P-value:", p)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(table, annot=True, fmt='d', cmap='coolwarm')
plt.title("Stress vs Mental Health")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X = df_clean[['stress', 'age', 'gender']]
y = df_clean['mental_health_binary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)

rf.fit(X_train, y_train)
```

```python
accuracy = rf.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
import pandas as pd

importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf.feature_importances_
})

importance = importance.sort_values(by='Importance', ascending=False)
importance
```

```python
import matplotlib.pyplot as plt

plt.barh(importance['Feature'], importance['Importance'])
plt.title("Feature Importance (Random Forest)")
plt.gca().invert_yaxis()
plt.show()
```
